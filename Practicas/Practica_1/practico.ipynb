```python
# Importar librerías necesarias
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
```

```python
# Cargar los datos
df = pd.read_csv('tabla.csv')
estaturas = df[['Estatura (m)']]
pesos = df['Peso (kg)']
```

```python
# Función para crear una curva (en este caso, una línea recta)
def line(x, a, b):
    return a * x + b
```

```python
# Visualizar los datos
plt.figure(figsize=(10, 6))
plt.scatter(df['Estatura (m)'], df['Peso (kg)'], alpha=0.5, label='Datos')
plt.xlabel('Estatura (m)')
plt.ylabel('Peso (kg)')
plt.title('Distribución de Peso vs Estatura')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Ajustar un modelo lineal
X = df[['Estatura (m)']]
y = df['Peso (kg)']

linear_model = LinearRegression()
linear_model.fit(X, y)
y_pred = linear_model.predict(X)
```

```python
# Visualizar el ajuste lineal
plt.figure(figsize=(10, 6))
plt.scatter(df['Estatura (m)'], df['Peso (kg)'], alpha=0.5, label='Datos')
plt.plot(df['Estatura (m)'], y_pred, color='red', label='Ajuste Lineal')
plt.xlabel('Estatura (m)')
plt.ylabel('Peso (kg)')
plt.title('Ajuste Lineal a los Datos')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Ajustar un modelo polinómico (grado 2)
poly = PolynomialFeatures(degree=2)
X_poly = poly.fit_transform(X)

poly_model = LinearRegression()
poly_model.fit(X_poly, y)
y_poly_pred = poly_model.predict(X_poly)
```

```python
# Visualizar el ajuste polinómico
plt.figure(figsize=(10, 6))
plt.scatter(df['Estatura (m)'], df['Peso (kg)'], alpha=0.5, label='Datos')
plt.plot(df['Estatura (m)'], y_poly_pred, color='green', label='Ajuste Polinómico (grado 2)')
plt.xlabel('Estatura (m)')
plt.ylabel('Peso (kg)')
plt.title('Ajuste Polinómico a los Datos')
plt.legend()
plt.grid(True)
plt.show()
```
